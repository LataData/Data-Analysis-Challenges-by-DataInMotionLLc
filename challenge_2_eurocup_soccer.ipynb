**Sort the teams by Red Cards, then to Yellow Cards.**

```python
df=df.sort_values(by=['Team','Yellow Cards','Red Cards'],ascending=[True, True,True])
df.head(2)
```

**Calculate the mean Yellow Cards given per Team.**

```python
df.groupby('Team')['Yellow Cards'].mean()
```

**Filter teams that scored more than 6 goals.**

**Germany  and Spain **

```python
df[df['Goals']>6][['Team','Goals']]
```

**Select the teams that start with the letter G.**

**Germany , Greece**

```python
df[df['Team'].str.startswith('G')]['Team']
```
